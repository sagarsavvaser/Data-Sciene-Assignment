```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):  
        if num % i == 0:
            return False
    return True


number = int(input("Enter a number: "))

if is_prime(number):
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

```python
import random


num1 = random.randint(1, 10)
num2 = random.randint(1, 10)


print(f"What is {num1} × {num2}?")
user_answer = int(input("Your answer: "))


correct_answer = num1 * num2

if user_answer == correct_answer:
    print("✅ Correct! Well done.")
else:
    print(f"❌ Incorrect. The correct answer is {correct_answer}.")
```

```python

room_charge_per_day = 2000
consultation_fee = 1500
lab_test_charge = 300
discount_threshold = 10000
discount_rate = 0.10  # 10%


days_stayed = 4
num_tests = int(input("Enter number of lab tests taken: "))
medicine_cost = float(input("Enter total medicine cost (₹): "))


room_cost = days_stayed * room_charge_per_day
lab_cost = num_tests * lab_test_charge
total_before_discount = room_cost + consultation_fee + lab_cost + medicine_cost


if total_before_discount > discount_threshold:
    discount = total_before_discount * discount_rate
else:
    discount = 0

final_bill = total_before_discount - discount

# Output
print("\n------ Hospital Bill ------")
print(f"Room charges (₹2000 × {days_stayed} days): ₹{room_cost}")
print(f"Doctor consultation fee: ₹{consultation_fee}")
print(f"Lab test charges (₹300 × {num_tests}): ₹{lab_cost}")
print(f"Medicine charges: ₹{medicine_cost}")
print(f"Total before discount: ₹{total_before_discount}")
print(f"Discount: ₹{discount}")
print(f"Final Bill: ₹{final_bill}")
```
